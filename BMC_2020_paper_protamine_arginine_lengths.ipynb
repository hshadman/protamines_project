```python
#protamine_sequence='MARYRCCRSRSRCRRRRRRCYRRRRRCCRRRRRRVCCRRYTVIRCRRR'
#look at example in above line. protamine_sequence is a single sequence. 
def count_arg_blocks(protamine_sequence):
    import numpy as np
    arg_length=[]
    i=0
    j=0
    k=1
    while j<=len(protamine_sequence)-2 and k<=len(protamine_sequence):
        if protamine_sequence[j]=='R' and protamine_sequence[j+1]=='R':
            count_arg=0
            i=j
            while protamine_sequence[i]=='R':
                count_arg=count_arg+1
                i=i+1
                if i>(len(protamine_sequence)-1):
                        break
            arg_length.extend([count_arg]*count_arg)
            j=i
        elif protamine_sequence[j]=='R' and protamine_sequence[j+1]!='R':
            arg_length.append(1)
            j=j+1
        elif protamine_sequence[j]=='-':
            arg_length.append(np.nan)
            j=j+1
        else:
            arg_length.append(0)
            j=j+1
        k=k+1
    if protamine_sequence[len(protamine_sequence)-1]=='R' and protamine_sequence[len(protamine_sequence)-2]!='R':
        arg_length.append(1)
#    elif protamine_sequence[len(protamine_sequence)-1]=='-' and protamine_sequence[len(protamine_sequence)-2]!='R':
#        arg_length.append(np.nan)
#    elif protamine_sequence[len(protamine_sequence)-1]=='-' and protamine_sequence[len(protamine_sequence)-2]=='R':
#        arg_length.append(np.nan)
#    elif protamine_sequence[len(protamine_sequence)-1]!='R' and protamine_sequence[len(protamine_sequence)-2]!='R' and protamine_sequence[len(protamine_sequence)-1]!='-':
#        arg_length.append(0)
    elif protamine_sequence[len(protamine_sequence)-1]=='-':
        arg_length.append(np.nan)
    elif protamine_sequence[len(protamine_sequence)-1]!='R':
        arg_length.append(0)
    return arg_length
```

```python
#accepts whole group of aligned sequences
#uses a previously defined function
def block_vs_single(protamine_sequence):
    import math
    counter=0
    for i in protamine_sequence:
        temp_sequence=count_arg_blocks(''.join(i))
        binary_protamine_temp=np.empty(shape=[0,0])
        for j in temp_sequence:
            if math.isnan(j)!=True and j<4:
                binary_protamine_temp=np.append(binary_protamine_temp,1)
            elif math.isnan(j)!=True and j>=4:
                binary_protamine_temp=np.append(binary_protamine_temp,5)
            elif math.isnan(j):
                binary_protamine_temp=np.append(binary_protamine_temp,np.nan)
        if counter==0:
            binary_protamine=binary_protamine_temp
        elif counter>0:
            binary_protamine=np.vstack([binary_protamine,binary_protamine_temp])
        counter+=1
    return binary_protamine
```

```python
#counts frequency of groups of arginines of 4 or more vs single arginines (3 or lesss) in aligned sequences
def sequence_to_binary_blockcount(protamine_sequence):
    import math
    binary_protamine_single=np.zeros(shape=(len(protamine_sequence[0]),))
    binary_protamine_group=np.zeros(shape=(len(protamine_sequence[0]),))
    for i in block_vs_single(protamine_sequence):
        for index,j in enumerate(i):
            if math.isnan(j)!=True and j==1:
                np.put(binary_protamine_single,index,binary_protamine_single[index]+1)
            elif math.isnan(j)!=True and j==5:
                np.put(binary_protamine_group,index,binary_protamine_group[index]+1)
            elif math.isnan(j):
                pass
    return binary_protamine_single,binary_protamine_group
```

```python
#functions that apply to aligned sequences from the folder obtained from BMC paper supplementary
def sequence_to_binary(protamine_sequence):
    counter=0
    for i in protamine_sequence:
        binary_protamine_temp=np.empty(shape=[0,0])
        for j in i:     
            if j=='R':
                binary_protamine_temp=np.append(binary_protamine_temp,1)
            else:
                binary_protamine_temp=np.append(binary_protamine_temp,0)
        if counter==0:
            binary_protamine=binary_protamine_temp
        elif counter>0:
            binary_protamine=np.vstack([binary_protamine,binary_protamine_temp])
        counter+=1
    return binary_protamine
#functions that apply to aligned sequences from the folder obtained from BMC paper supplementary
def hydrophobic_to_binary(protamine_sequence):
    counter=0
    for i in protamine_sequence:
        binary_protamine_temp=np.empty(shape=[0,0])
        for j in i:     
            if j in ['A','I','L','M','P','V','Y','W','F','G','T','Q']:
                binary_protamine_temp=np.append(binary_protamine_temp,1)
            else:
                binary_protamine_temp=np.append(binary_protamine_temp,0)
        if counter==0:
            binary_protamine=binary_protamine_temp
        elif counter>0:
            binary_protamine=np.vstack([binary_protamine,binary_protamine_temp])
        counter+=1
    return binary_protamine

#charge: arginine and lysine converted to +1, glutamic acid and aspartic acid converted to -1
def sequence_to_charge(protamine_sequence):
#ignoring histidine
    counter=0
    for i in protamine_sequence:
        binary_protamine_temp=np.empty(shape=[0,0])
        for j in i:     
            if j=='R' or j=='K':
                binary_protamine_temp=np.append(binary_protamine_temp,1)
            elif j=='D' or j=='E':
                binary_protamine_temp=np.append(binary_protamine_temp,-1)
            else:
                binary_protamine_temp=np.append(binary_protamine_temp,0)
        if counter==0:
            binary_protamine=binary_protamine_temp
        elif counter>0:
            binary_protamine=np.vstack([binary_protamine,binary_protamine_temp])
        counter+=1
    return binary_protamine
def sequence_to_binary_dashnan(protamine_sequence):
    counter=0
    for i in protamine_sequence:
        binary_protamine_temp=np.empty(shape=[0,0])
        for j in i:     
            if j=='R':
                binary_protamine_temp=np.append(binary_protamine_temp,1)
            elif j=='-':
                binary_protamine_temp=np.append(binary_protamine_temp,np.nan)
            else:
                binary_protamine_temp=np.append(binary_protamine_temp,0)
        if counter==0:
            binary_protamine=binary_protamine_temp
        elif counter>0:
            binary_protamine=np.vstack([binary_protamine,binary_protamine_temp])
        counter+=1
    return binary_protamine
def sequence_to_binary_dashcount(protamine_sequence):
    binary_protamine_temp=np.zeros(shape=(len(protamine_sequence[0]),))
    for i in protamine_sequence:
        for index,j in enumerate(i):
            if j=='-':
                np.put(binary_protamine_temp,index,binary_protamine_temp[index]+1)
    return binary_protamine_temp
```

```python
#this one takes moving window and calculates average of non_nan values in that window
#if all values are nan in the window then a 0 is recorded to be the average of that window
def custom_uniform_filter_1d_wave(mylist,N):
    nan_amend=list()
    temp_list=[x for x in mylist if math.isnan(x)==False]
    for i in range(1,(N//2)+1):
        if N%2==1:
            mylist.insert(0,temp_list[i-1])
            mylist.append(temp_list[len(temp_list)-i])
        elif N%2==0:
            if i<(N//2):
                mylist.insert(0,temp_list[i-1])
                mylist.append(temp_list[len(temp_list)-i])
            elif i==N//2:
                mylist.insert(0,temp_list[i-1])
    moving_aves = []
    start=0
    for i,x in enumerate(mylist):
        if start+N<=len(mylist):
            moving_window=mylist[start:(start+N)]
            moving_window_nonan=[x for x in moving_window if math.isnan(x)==False]
            if len(moving_window_nonan)==0:
                moving_aves.append(0)
            else:
                moving_aves.append(sum(moving_window_nonan)/len(moving_window_nonan))
        start=start+1
    return moving_aves
```
