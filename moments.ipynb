```python
salmon_tensor['Rg_sqrd']=salmon_tensor.R1.values+salmon_tensor.R2.values+salmon_tensor.R3.values
salmon_tensor['asphericity']=salmon_tensor.R1.values-(0.5*(salmon_tensor.R2.values+salmon_tensor.R3.values))
salmon_tensor['acylindricity']=salmon_tensor.R2.values-salmon_tensor.R3.values
salmon_tensor['RSA']=((salmon_tensor.asphericity.values**2+(0.75*salmon_tensor.acylindricity.values**2))/(salmon_tensor.Rg_sqrd.values)**2)**0.5
salmon_tensor['ratio']=salmon_tensor.Ree.values**2/salmon_tensor.Rg_sqrd.values
p1_tensor['Rg_sqrd']=p1_tensor.R1.values+p1_tensor.R2.values+p1_tensor.R3.values
p1_tensor['asphericity']=p1_tensor.R1.values-(0.5*(p1_tensor.R2.values+p1_tensor.R3.values))
p1_tensor['acylindricity']=p1_tensor.R2.values-p1_tensor.R3.values
p1_tensor['RSA']=((p1_tensor.asphericity.values**2+(0.75*p1_tensor.acylindricity.values**2))/(p1_tensor.Rg_sqrd.values)**2)**0.5
p1_tensor['ratio']=p1_tensor.Ree.values**2/p1_tensor.Rg_sqrd.values
bull_tensor['Rg_sqrd']=bull_tensor.R1.values+bull_tensor.R2.values+bull_tensor.R3.values
bull_tensor['asphericity']=bull_tensor.R1.values-(0.5*(bull_tensor.R2.values+bull_tensor.R3.values))
bull_tensor['acylindricity']=bull_tensor.R2.values-bull_tensor.R3.values
bull_tensor['RSA']=((bull_tensor.asphericity.values**2+(0.75*bull_tensor.acylindricity.values**2))/(bull_tensor.Rg_sqrd.values)**2)**0.5
bull_tensor['ratio']=bull_tensor.Ree.values**2/bull_tensor.Rg_sqrd.values
```
